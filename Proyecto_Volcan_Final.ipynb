# Erupciones Volcánicas

**¿Qué es una erupción volcánica?**

Una erupción volcánica es un fenómeno geológico que ocurre cuando el magma (roca fundida) y los gases acumulados en el interior de la Tierra ascienden a la superficie a través de una abertura en la corteza terrestre, comúnmente conocido como volcán. Durante este proceso, se liberan gases, cenizas, rocas y lava.

Las erupciones volcánicas son el resultado de la acumulación de presión en el interior de la Tierra. El magma, más caliente y menos denso que las rocas circundantes, tiende a ascender. Cuando encuentra una abertura, sale violentamente, generando la erupción.

El **Índice de Explosividad Volcánica (VEI)** es una escala que se utiliza para medir la magnitud y violencia de una erupción volcánica. Se basa en varios factores, como la altura de la columna eruptiva, el volumen de material expulsado y la duración de la erupción.

* **Escala logarítmica:** El VEI es una escala logarítmica, lo que significa que cada incremento de un número representa una erupción diez veces más energética.
* **Rango:** Va desde el 0 (erupciones muy pequeñas) hasta el 8 (erupciones extremadamente grandes y catastróficas).

**Tipos de Volcanes**

Los volcanes se clasifican según su forma, tipo de erupción y composición de la lava. Algunos de los tipos más comunes son:

* **Volcanes en escudo:** Tienen forma de cono bajo y pendientes suaves. Producen erupciones efusivas de lava fluida, como los volcanes de Hawái.
* **Estratovolcanes:** Tienen forma cónica y pendientes empinadas. Producen erupciones explosivas de lava viscosa y materiales piroclásticos (cenizas, rocas).
* **Volcanes de caldera:** Se forman cuando el techo de una cámara magmática colapsa, creando una gran depresión.
* **Volcanes de lodo:** Expulsan principalmente lodo y gases, en lugar de lava.

Finalmente la **configuración tectónica** se refiere a la disposición de las placas tectónicas y los procesos geológicos que ocurren en sus bordes. La actividad volcánica está estrechamente relacionada con la tectónica de placas:

* **Zonas de subducción:** En estas zonas, una placa tectónica se sumerge por debajo de otra. La fricción y el calor generan magma, que puede ascender y formar volcanes en arcos insulares o cordilleras continentales.
* **Dorsales oceánicas:** En estas zonas, las placas tectónicas se separan, permitiendo que el magma ascienda y forme nueva corteza oceánica.
* **Puntos calientes:** Son zonas del manto terrestre donde se producen plumas de material caliente que ascienden hacia la superficie, generando volcanes en el interior de las placas tectónicas.

En la siguiente se calculará el tiempo transcurrido en días en las que la erupción duró.

```python
# Combinar las columnas de fecha para formar una fecha completa
erupciones['start_date'] = pd.to_datetime(
    erupciones[['start_year', 'start_month', 'start_day']].rename(
        columns={'start_year': 'year', 'start_month': 'month', 'start_day': 'day'}
    ),
    errors='coerce'
)

erupciones['end_date'] = pd.to_datetime(
    erupciones[['end_year', 'end_month', 'end_day']].rename(
        columns={'end_year': 'year', 'end_month': 'month', 'end_day': 'day'}
    ),
    errors='coerce'
)

# Calcular la duración (en días)
erupciones['eruption_duration'] = (erupciones['end_date'] - erupciones['start_date']).dt.days + 1

# Calcular el primer cuartil (Q1) de la columna 'eruption_duration', excluyendo los NaN
q1_duration = erupciones['eruption_duration'].dropna().quantile(0.25)

# Rellenar los valores faltantes con el primer cuartil (Q1)
erupciones['eruption_duration'].fillna(q1_duration, inplace=True)

# Convertir a un formato entero para claridad
erupciones['eruption_duration'] = erupciones['eruption_duration'].astype(int)

# Visualización
display(erupciones.head())
```

```python
# Calcular el promedio de 'vei' por medio de la duración de la erupción (posdata sé que eso no es tan correlacionado, pero ajá...)
# Ignorando los valores nulos
vei_mean_by_year = erupciones.groupby('eruption_duration')['vei'].mean()

# Rellenar los valores faltantes de 'vei' con el promedio correspondiente al año de inicio
erupciones['vei'] = erupciones.apply(
    lambda row: vei_mean_by_year[row['eruption_duration']] if pd.isnull(row['vei']) else row['vei'], axis=1
)

# Si todavía quedan valores nulos (años sin datos), rellenarlos con el promedio general
general_mean_vei = erupciones['vei'].mean()
erupciones['vei'].fillna(general_mean_vei, inplace=True)

# Verificar la limpieza
erupciones['vei'].isnull().sum()
display(erupciones.head())
```

```python
# Redondear la columna 'vei' a enteros
erupciones['vei'] = np.round(erupciones['vei']).astype(int)

# Verificar los cambios
display(erupciones.head())
```

```python
# Función para detectar outliers usando IQR
def detect_outliers(df, column):
  Q1 = df[column].quantile(0.25)
  Q3 = df[column].quantile(0.75)
  IQR = Q3 - Q1
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR
  return df[(df[column] < lower_bound) | (df[column] > upper_bound)]

# Detectar outliers en 'vei', 'eruption_duration' y 'eruptions_number'
outliers_vei = detect_outliers(union_1, 'vei')
outliers_eruption_duration = detect_outliers(union_1, 'eruption_duration')
outliers_n_eruptions = detect_outliers(union_1, 'eruption_number_y')
```

# Machine Learning

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Seleccionamos las columnas relevantes
features = union_1[['latitude', 'longitude', 'eruption_category', 'start_year', 'vei', 'eruption_duration']]
target = union_1['eruption_duration']  # Predecimos la duración de la erupción

# Convertir 'eruption_category' a variable numérica
features = pd.get_dummies(features, columns=['eruption_category'], drop_first=True)

# Dividir los datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
# Entrenamiento de un modelo Random Forest Regressor
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Predicciones
y_pred = model.predict(X_test)

# Evaluación del modelo
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error (MSE): {mse}')
print(f'R2 Score: {r2}')
```
